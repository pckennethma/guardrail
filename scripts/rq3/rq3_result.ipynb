```python
dataset_name_ls = [
    "adult",
    "lung_cancer",
    "cylinder-bands",
    "diabetes",
    "contraceptive",
    "blood-transfusion-service-center",
    "steel-plates-fault",
    "jungle_chess_2pcs_raw_endgame_complete",
    "telco_customer_churn",
    "bank-marketing",
    "PhishingWebsites",
    "hotel_bookings",
]
short_name_ls = [
    "Adult",
    "Lung Cancer",
    "Cylinder Bands",
    "Diabetes",
    "Contraceptive Method",
    "Blood Trans. Serv. Ctr.",
    "Steel Plates Faults",
    "Jungle Chess",
    "Telco Customer Churn",
    "Bank Marketing",
    "Phishing Websites",
    "Hotel Reservations",
]

model_dir = "../models/"
output_str = ""
out_put_str_len = ""
coverage_identify = []
coverage_aux = []
for i in range(len(dataset_name_ls)):
    for filename in os.listdir(model_dir):
        if filename.startswith(dataset_name_ls[i]):
            model_folder = filename
            break

    # get length of the dataset
    df_path = _DATASET_PATH.get(dataset_name_ls[i], None)
    df_path = df_path.replace(".csv", ".train.csv")
    df_path = os.path.join("..", df_path)
    df = pd.read_csv(df_path)

    with open(
        os.path.join(
            model_dir, model_folder, "sampler", f"nsyn_prog_sampler_use_aux_0_idx_0.pkl"
        ),
        "rb",
    ) as f:
        data_log = pickle.load(f)
        coverage_identify.append(data_log.coverage / len(df))
        len1 = len(data_log.stmts)

    with open(os.path.join(model_dir, model_folder, f"nsyn_prog.pkl"), "rb") as f:
        data_log = pickle.load(f)
        coverage_aux.append(data_log.coverage / len(df))
        len2 = len(data_log.stmts)

    coverage_identify[-1] = coverage_identify[-1] / max(len1, len2)
    coverage_aux[-1] = coverage_aux[-1] / max(len1, len2)

fig, ax = plt.subplots(figsize=(5, 6))
ax.set_yticks(len(short_name_ls) - np.arange(len(short_name_ls)), labels=short_name_ls)
ax.set_xlabel("Coverage (Normalized)")
ax.set_xlim(-0.05, 1.05)
ax.grid()


ax.scatter(
    coverage_identify,
    [len(short_name_ls) - i for i in range(len(short_name_ls))],
    color="b",
    label="Identity Sampler",
    zorder=2,
    s=100,
)
ax.scatter(
    coverage_aux,
    [len(short_name_ls) - i for i in range(len(short_name_ls))],
    color="r",
    label="Auxiliary Sampler",
    zorder=2,
    s=100,
)

handles, labels = ax.get_legend_handles_labels()
fig.legend(
    handles, labels, loc="center", bbox_to_anchor=(0.5, 0.0), ncol=2, fontsize=11
)
plt.show()
fig.savefig("rq3_sampler.pdf", bbox_inches="tight")
```

```python
def get_coverage_loss(dataset_name):
    model_dir = "../models/"
    for filename in os.listdir(model_dir):
        if filename.startswith(dataset_name):
            model_folder = filename
            break

    df_path = _DATASET_PATH.get(dataset_name, None)
    df_path = df_path.replace(".csv", ".train.csv")
    df_path = os.path.join("..", df_path)
    df = pd.read_csv(df_path)

    coverage_ls = []
    loss_ls = []

    for epsilon in [0.01, 0.05, 0.1, 0.15, 0.2, 0.25]:
        with open(
            os.path.join(
                model_dir,
                model_folder,
                "epsilon",
                f"nsyn_prog_epsilon_{epsilon}_idx_0_with_loss.pkl",
            ),
            "rb",
        ) as f:
            data_log = pickle.load(f)
            coverage_ls.append(data_log.coverage / len(df))
            loss_ls.append(
                sum(stmt.loss for stmt in data_log.stmts) / len(df)
                if data_log.stmts
                else 0
            )

    coverage_ls = [i / max(coverage_ls) for i in coverage_ls]
    loss_ls = [i / max(loss_ls) for i in loss_ls]
    return coverage_ls, loss_ls
```
